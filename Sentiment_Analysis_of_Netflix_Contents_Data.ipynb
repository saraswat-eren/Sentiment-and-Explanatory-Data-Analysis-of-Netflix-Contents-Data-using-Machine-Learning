```python
df['director'] = df['director'].fillna('No Director Specified')
df.head()
```

```python
directors_list = pd.DataFrame()
directors_list = df['director'].str.split(',', expand=True).stack()
print(directors_list)
```

```python
directors_list = directors_list.to_frame()
print(directors_list)
```

```python
directors_list.columns = ['directors']
print(directors_list)
```

```python
directors = directors_list.groupby(['directors']).size().reset_index(name='counts')
print(directors)
```

```python
directors = directors[directors.directors != 'No Director Specified']
print(directors)
```

```python
directors = directors.sort_values(by=['counts'], ascending=False)
print(directors)
```

```python
top5_directors = directors.head()
print(top5_directors)
```

```python
BarChart = px.bar(top5_directors, x='directors', y='counts', title='Top 5 Directors on Netflix')
print(BarChart)
```

```python
BarChart.show()
```

```python
# Analyzing the content produced on Netflix based on years

df1 = df[['type', 'release_year']]
df1 = df1.rename(columns={"release_year":"Release Year"})
df2 = df1.groupby(['Release Year', 'type']).size().reset_index(name='counts')
print(df2)
```

```python
graph = px.line(df2, x='Release Year', y='counts', title='Netflix Content produced over the years')
print(graph)
```

```python
graph.show()
```

Sentiment Analysis of Netflix Content

```python
df3 = df[['release_year', 'description']]
df3 = df3.rename(columns={"release_year":"Release Year"})
for index, row in df3.iterrows():
  d=row['description']
  s=TextBlob(d)
  sentiment=s.sentiment.polarity
  if sentiment==0:
    sent = 'Neutral'
  elif sentiment>0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df3.loc[[index, 2], 'Sentiment'] = sent

df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='counts')


print(df3)
```

```python
BarGraph = px.bar(df3, x='Release Year', y='counts', color='Sentiment', title='Sentiment Analysis of Netflix Content')
print(BarGraph)
```

```python
BarGraph.show()
```
